#Creating a dataframe 

```python
data = {'Date': [None,'2023-06-13', '2022-12-02', '2012-03-02', '2001-04-01', '2022-04-23','2022-04-23', '2013-04-23', '2022-12-02'], 'Age': [23,21,44,34,None,34, 34, 74, 74], 'First_Name':['Becky','Becky', 'Ali', 'Heuijing', None, 'Bob','Bob', 'Francoise', 'Ali'],'Last_name':['Smith', None, 'Phillips', 'Middleton', 'McAvoy', 'Khan', 'Cho', 'Smith', 'Smith'], 'Colour':['Blue','Cyan', 'Green', 'Yellow', 'Burgundy', 'Magnolia', 'Emerald','Teal', 'Teal'], 'ID':['Bec01','Bec02', 'Ali01', 'Heu04','Anna01' ,'Bob01', 'Bob02', 'Francoise01', 'Ali01'], 'Stage':['Stage_1', None, 'Stage_1','Stage01','Stage_1','Stage_2','Stage_2', 'Stage_2', 'Stage_1'], 'Work': ['desk-accountant', 'phyical-plumber', 'physical-builder','desk-manager', 'desk-lawyer', 'desk-engineer', 'desk-banker', 'desk-banker', 'desk_banker']}
data_df = pd.DataFrame(data, columns = ['Date', 'Age', 'First_Name','Last_name', 'Colour', 'ID', 'Stage', 'Work'])
data_df
```

```python
#identifying and quantifying unique entries
print('Unique Stage allocations:')
print(data_df['Stage'].unique())
print('*****')
print('Quanitiy of unique stage allocations:')
print(data_df['Stage'].value_counts())
print('******')
#unique groupby entries 
print('groupby name to find number of stages')
print(data_df.groupby(['Stage']).apply(lambda x: x['First_Name']).value_counts())
print('******')
print('groupby name and id to find the number of stages')
print(data_df.groupby(['First_Name', 'ID']).apply(lambda x: x['Stage'].value_counts()))
```
